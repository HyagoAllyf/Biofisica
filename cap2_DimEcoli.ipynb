```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Carrega a imagem em escala de cinza
graticule = cv2.imread('Graticule100x.tif', cv2.IMREAD_GRAYSCALE)

# Exibe informações sobre a imagem (opcional)
print(graticule.shape)  # Imprime a forma da matriz da imagem

# Plota a imagem
plt.figure(1)
plt.imshow(graticule, cmap='gray')  # Mostra a imagem em escala de cinza
plt.axis('off')  # Desativa os eixos
plt.show()
```

```python
# Carrega a imagem
graticula = cv2.imread('Graticule100x.tif', cv2.IMREAD_GRAYSCALE)

# Plota a linha
plt.figure(2)
linha = 400
plot_handle, = plt.plot(graticula[linha, :], '-k')
plt.xlim([1, graticula.shape[1]])  # Define o limite x
plt.xlabel('pixel')
plt.ylabel('intensidade')

# Espera pelo input de dois cliques
clicks = plt.ginput(2) # aprender como usar

# Calcula a diferença entre os cliques
diferenca_x = clicks[1][0] - clicks[0][0]

print("Diferença em x:", diferenca_x)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Carrega a imagem E. coli
Ecoli = cv2.imread('Ecoli100x.tif', cv2.IMREAD_GRAYSCALE)

# Normaliza a imagem entre 0 e 1
EcoliNorm = cv2.normalize(Ecoli, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

# Exibe a imagem normalizada
plt.figure(3)
plt.imshow(EcoliNorm, cmap='gray')
plt.axis('off')
plt.show()

# Calcula 2000/64.7
resultado = 2000 / 64.7
print("Resultado:", resultado)

# Escolhe uma seção da imagem e a transforma em zeros
EcoliNorm[900:906, 1200:1232] = 0

# Exibe a imagem com a seção transformada em zeros
plt.figure(4)
plt.imshow(EcoliNorm, cmap='gray')
plt.axis('off')
plt.show()

# Obtém o tamanho da imagem EcoliNorm
tamanho = EcoliNorm.shape
print("Tamanho da imagem EcoliNorm:", tamanho)

# Salva a imagem modificada
cv2.imwrite('EcoliCalibrado.tif', EcoliNorm * 255)
```
